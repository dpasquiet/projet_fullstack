```python
# Regrouper par 'Filière de formation.1' et calculer la somme des autres colonnes
df_grouped = df.groupby('Filière de formation.1').sum().reset_index()
```

```python
df_grouped.head()
```

```python
df_grouped.shape
```

```python
df_grouped['Session'] = 2023
```

```python
# Réorganiser les colonnes pour placer 'Session' en première position
colonnes = ['Session'] + [colonne for colonne in df_grouped.columns if colonne != 'Session']
df_grouped = df_grouped[colonnes]
```

```python
df_grouped.head()
```

```python
# Liste des valeurs à supprimer
valeurs_a_supprimer = [
"Année préparatoire",
'BPJEPS',
'BTS - Maritime',
'C.M.I - Cursus Master en Ingénierie',
'Classe préparatoire aux études supérieures',
"Diplôme d'Etablissement",
"Diplôme d'établissement",
"Diplôme d'Université",
'FCIL',
'Formation des écoles supérieures de cuisine',
'Formation professionnelle',
'Formation valant grade de licence',
'Formations Bac + 3',
'Formations Bac + 5',
'LP -  Sciences Humaines et Sociales',
'LP - Droit-économie-gestion',
'LP - Sciences - technologies - santé',
'Mention complémentaire',
'Mise à niveau',
'DEJEPS',
'CPES'
]

# Supprimer les lignes correspondant aux valeurs dans la liste
df_filtre = df_grouped[~df_grouped['Filière de formation.1'].isin(valeurs_a_supprimer)]
```

```python
df_filtre.shape
```

```python
# Filtrer les lignes où la colonne 'Nom' contient 'CUPGE'
df_filtre[df_filtre['Filière de formation.1'].str.contains('CUPGE', case=False, na=False)]
```

```python
# Filtrer les lignes où la colonne 'Filière de formation.1' contient 'CUPGE'
df_cupge = df_filtre[df_filtre['Filière de formation.1'].str.contains('CUPGE', case=False, na=False)]

# Remplacer les valeurs de 'Filière de formation.1' par 'CUPGE' pour le regroupement
df_cupge['Filière de formation.1'] = 'CUPGE'

# Regrouper par 'Filière de formation.1' et calculer la somme des autres colonnes
df_cupge_regroupe = df_cupge.groupby('Filière de formation.1').sum().reset_index()

# Filtrer les lignes où la colonne 'Filière de formation.1' ne contient pas 'CUPGE'
df_autres = df_filtre[~df_filtre['Filière de formation.1'].str.contains('CUPGE', case=False, na=False)]

# Combiner les deux DataFrames
df_final = pd.concat([df_cupge_regroupe, df_autres], ignore_index=True)
```

```python
df_final.shape
```

```python
df_final.loc[0, 'Session'] = 2023
```

```python
df_final.head()
```

```python
# Liste des valeurs à regrouper
valeurs_a_regrouper = ['DN MADE', 'Diplôme National d\'Art', 'Formation des écoles supérieure d\'art']

# Filtrer les lignes où la colonne 'Filière de formation.1' contient les valeurs spécifiées
df_art = df_final[df_final['Filière de formation.1'].isin(valeurs_a_regrouper)]

# Remplacer les valeurs de 'Filière de formation.1' par 'Formations d\'ART' pour le regroupement
df_art['Filière de formation.1'] = 'Formations d\'ART'

# Regrouper par 'Filière de formation.1' et calculer la somme des autres colonnes
df_art_regroupe = df_art.groupby('Filière de formation.1').sum().reset_index()

# Filtrer les lignes où la colonne 'Filière de formation.1' ne contient pas les valeurs spécifiées
df_autres = df_final[~df_final['Filière de formation.1'].isin(valeurs_a_regrouper)]

# Combiner les deux DataFrames
df_final2 = pd.concat([df_art_regroupe, df_autres], ignore_index=True)
```

```python
df_final2.head()
```

```python
df_final2.loc[0, 'Session'] = 2023
```

```python
df_final2.head()
```

```python
# Réorganiser les colonnes pour placer 'Session' en première position
colonnes = ['Session'] + [colonne for colonne in df_final2.columns if colonne != 'Session']
df_final2_grouped = df_final2[colonnes]
```
